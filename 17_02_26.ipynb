```python
import pandas as pd

# Read dataset from URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("First 5 rows of dataset:")
print(df.head())


# --------------------------------------------------
# a) DATA FILTERING
# --------------------------------------------------

# Filter passengers who survived and are female
filtered_data = df[(df['Survived'] == 1) & (df['Sex'] == 'female')]

print("\nFiltered Data (Survived Females):")
print(filtered_data.head())


# --------------------------------------------------
# b) DATA MANIPULATION
# --------------------------------------------------

# Create a new column: Family Size
df['Family_Size'] = df['SibSp'] + df['Parch']

# Fill missing Age values with mean age
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Drop Cabin column (many missing values)
df.drop(columns=['Cabin'], inplace=True)

print("\nData after Manipulation:")
print(df.head())


# --------------------------------------------------
# c) DATA RANKING
# --------------------------------------------------

# Rank passengers based on Fare (highest fare = rank 1)
df['Fare_Rank'] = df['Fare'].rank(ascending=False)

print("\nTop 5 passengers by Fare:")
print(df[['Name', 'Fare', 'Fare_Rank']].sort_values(by='Fare_Rank').head())
```

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {
    'Marks': [78, 85, 96, 65, 70, 85, 90, 88, 76, 95]
}

df = pd.DataFrame(data)

print("Data Frame:")
print(df)
# a) Mean, Median, Mode

print("\nMean:", df['Marks'].mean())
print("Median:", df['Marks'].median())
print("Mode:")
print(df['Marks'].mode())

# b) Standard Deviation

print("\nStandard Deviation:", df['Marks'].std())

# c) Variance
print("Variance:", df['Marks'].var())

# d) Kurtosis
print("Kurtosis:", df['Marks'].kurt())

# e) Skewness
print("Skewness:", df['Marks'].skew())
```

```python
# Import library
import pandas as pd

# Read dataset from URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("First 5 rows of dataset:")
print(df.head())


# Select only numeric columns for statistics
numeric_df = df.select_dtypes(include=['int64', 'float64'])

print("\nDescriptive Statistics Summary:")
print(numeric_df.describe())

# Individual Descriptive Statistics
# Mean
print("\nMean:")
print(numeric_df.mean())

# Median
print("\nMedian:")
print(numeric_df.median())

# Mode
print("\nMode:")
print(numeric_df.mode().iloc[0])

# Standard Deviation
print("\nStandard Deviation:")
print(numeric_df.std())

# Variance
print("\nVariance:")
print(numeric_df.var())

# Skewness
print("\nSkewness:")
print(numeric_df.skew())

# Kurtosis
print("\nKurtosis:")
print(numeric_df.kurt())
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read dataset from URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("First 5 rows:")
print(df.head())

# 1. Histogram (Age Distribution)
plt.figure(figsize=(6,4))
plt.hist(df['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# 2. Bar Chart (Survival Count)
plt.figure(figsize=(6,4))
df['Survived'].value_counts().plot(kind='bar', color=['red','green'])
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# 3. Scatter Plot (Age vs Fare)
plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

# 4. Box Plot (Fare Distribution)
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")
plt.show()


# 5. Heatmap (Correlation Matrix)

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
