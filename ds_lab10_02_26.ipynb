```python

#selectionsort
# -------------------------------------------------
# Selection Sort Program in Python
# -------------------------------------------------
# This program sorts a list of numbers in ascending
# order using the Selection Sort algorithm.
# -------------------------------------------------

def selection_sort(arr):
    n = len(arr)

    # Traverse through all list elements
    for i in range(n):
        # Assume the current index has the minimum value
        min_index = i

        # Find the minimum element in remaining unsorted array
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        # Swap the found minimum element with the first element
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]

        # Display the list after each pass
        print(f"After pass {i + 1}: {arr}")

def main():
    # Read number of elements
    n = int(input("Enter number of elements: "))

    # Create empty list
    arr = []

    # Read elements from user
    for i in range(n):
        element = int(input(f"Enter element {i + 1}: "))
        arr.append(element)

    # Display original array
    print("\nOriginal array:", arr)

    # Perform selection sort
    print("\nSorting steps:")
    selection_sort(arr)

    # Display sorted array
    print("\nFinal sorted array:", arr)

# Call main function
main()
```

```python
#linearsearch
# -------------------------------------------------
# Linear Search Program in Python
# -------------------------------------------------
# This program searches for an element in a list
# using the Linear Search technique.
# -------------------------------------------------

def linear_search(arr, key):
    """
    This function performs linear search.
    It returns the index of the element if found,
    otherwise returns -1.
    """
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1


def main():
    # Read number of elements
    n = int(input("Enter number of elements: "))

    # Create empty list
    arr = []

    # Read elements into the list
    for i in range(n):
        element = int(input(f"Enter element {i + 1}: "))
        arr.append(element)

    # Display the list
    print("\nList elements:", arr)

    # Read element to search
    key = int(input("\nEnter element to search: "))

    # Perform linear search
    result = linear_search(arr, key)

    # Display result
    if result != -1:
        print(f"Element {key} found at position {result + 1}")
    else:
        print(f"Element {key} not found in the list")


# Call main function
main()
```

```python
#binarysearch
# -------------------------------------------------
# Binary Search Program in Python
# -------------------------------------------------
# This program searches for an element in a list
# using the Binary Search technique.
# -------------------------------------------------

def binary_search(arr, key):
    """
    This function performs binary search.
    It returns the index of the element if found,
    otherwise returns -1.
    """
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1


def main():
    # Read number of elements
    n = int(input("Enter number of elements: "))

    # Create empty list
    arr = []

    # Read elements into the list
    for i in range(n):
        element = int(input(f"Enter element {i + 1}: "))
        arr.append(element)

    # Sort the list (IMPORTANT for binary search)
    arr.sort()

    # Display sorted list
    print("\nSorted list:", arr)

    # Read element to search
    key = int(input("\nEnter element to search: "))

    # Perform binary search
    result = binary_search(arr, key)

    # Display result
    if result != -1:
        print(f"Element {key} found at position {result + 1}")
    else:
        print(f"Element {key} not found in the list")


# Call main function
main()
```

```python
#bubblesort
# -------------------------------------------------
# Bubble Sort Program in Python
# -------------------------------------------------
# This program sorts a list of numbers in ascending
# order using the Bubble Sort algorithm.
# -------------------------------------------------

def bubble_sort(arr):
    """
    This function performs Bubble Sort on a list.
    It sorts the list in ascending order.
    """
    n = len(arr)

    # Traverse through all elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Swap if the element found is greater
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

        # Display the array after each pass
        print(f"After pass {i + 1}: {arr}")


def main():
    # Read number of elements
    n = int(input("Enter number of elements: "))

    # Create empty list
    arr = []

    # Read elements into the list
    for i in range(n):
        element = int(input(f"Enter element {i + 1}: "))
        arr.append(element)

    # Display original array
    print("\nOriginal array:", arr)

    # Perform Bubble Sort
    print("\nSorting steps:")
    bubble_sort(arr)

    # Display final sorted array
    print("\nSorted array:", arr)


# Call main function
main()
```

```python
#datamanipulation
# -------------------------------------------------
# Data Manipulation Program in Python
# -------------------------------------------------
# This program demonstrates various data manipulation
# tasks like creating, updating, filtering, and
# aggregating data using pandas.
# -------------------------------------------------

import pandas as pd

# Step 1: Create DataFrame
data = {
    "ID": [101, 102, 103, 104],
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [23, 25, 22, 24],
    "Department": ["HR", "IT", "Finance", "IT"],
    "Salary": [50000, 60000, 55000, 62000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Step 2: Access specific column
print("\nNames:\n", df["Name"])

# Step 3: Access specific row by index
print("\nRow at index 2:\n", df.loc[2])

# Step 4: Filter rows based on condition
print("\nEmployees in IT Department:\n", df[df["Department"] == "IT"])

# Step 5: Add new column
df["Bonus"] = [5000, 6000, 5500, 6200]
print("\nAfter adding Bonus column:\n", df)

# Step 6: Update column values
df["Salary"] = df["Salary"] + df["Bonus"]
print("\nAfter updating Salary with Bonus:\n", df)

# Step 7: Delete a column
df.drop("Bonus", axis=1, inplace=True)
print("\nAfter deleting Bonus column:\n", df)

# Step 8: Sorting data
print("\nSorted by Salary (descending):\n", df.sort_values(by="Salary", ascending=False))

# Step 9: Aggregate functions
print("\nAverage Age:", df["Age"].mean())
print("Maximum Salary:", df["Salary"].max())

# Step 10: Grouping data
grouped = df.groupby("Department")["Salary"].mean()
print("\nAverage Salary by Department:\n", grouped)
```

```python
#ranking
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Score": [85, 92, 75, 92, 60]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Rank the scores (highest score gets rank 1)
df["Rank"] = df["Score"].rank(method="min", ascending=False)

print("\nDataFrame with Ranks:\n", df)
# List of scores
scores = [85, 92, 75, 92, 60]

# Sort scores in descending order and remove duplicates for ranking
sorted_scores = sorted(list(set(scores)), reverse=True)

# Assign ranks
ranks = [sorted_scores.index(score) + 1 for score in scores]

# Display results
print("Scores:", scores)
print("Ranks:", ranks)
```

```python
#plottingdata
# -------------------------------------------------
# Plotting Program in Python using Matplotlib
# -------------------------------------------------
# This program demonstrates different types of plots
# including line, bar, scatter, and histogram.
# -------------------------------------------------

import matplotlib.pyplot as plt

# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [250, 300, 280, 350, 400]
expenses = [200, 250, 230, 300, 350]

# -------------------------
# 1️⃣ Line Plot
# -------------------------
plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker='o', linestyle='-', color='blue', label='Sales')
plt.plot(months, expenses, marker='s', linestyle='--', color='red', label='Expenses')
plt.title("Monthly Sales vs Expenses")
plt.xlabel("Months")
plt.ylabel("Amount ($)")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------
# 2️⃣ Bar Chart
# -------------------------
plt.figure(figsize=(8, 5))
plt.bar(months, sales, color='green', alpha=0.6, label='Sales')
plt.bar(months, expenses, color='orange', alpha=0.6, label='Expenses')
plt.title("Monthly Sales and Expenses")
plt.xlabel("Months")
plt.ylabel("Amount ($)")
plt.legend()
plt.show()

# -------------------------
# 3️⃣ Scatter Plot
# -------------------------
plt.figure(figsize=(8, 5))
plt.scatter(sales, expenses, color='purple', s=100)
plt.title("Sales vs Expenses Scatter Plot")
plt.xlabel("Sales ($)")
plt.ylabel("Expenses ($)")
plt.grid(True)
plt.show()

# -------------------------
# 4️⃣ Histogram
# -------------------------
ages = [22, 25, 27, 23, 24, 22, 26, 28, 25, 24, 23, 27]
plt.figure(figsize=(8, 5))
plt.hist(ages, bins=5, color='skyblue', edgecolor='black')
plt.title("Age Distribution Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
